# 第一步：到入库

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 第二步：导入数据集

```python
dataset = pd.read_csv('datasets/Social_Network_Ads.csv')
```

```python
dataset
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
X = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, -1].values
```

# 第三步：将数据集划分为训练集和测试集

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=0)
```

# 第四步：特征缩放

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

# 第五步：使用RandomForestClassidier对训练集进行训练

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=10,
                                    criterion='entropy',
                                    random_state=0)
classifier.fit(X_train, y_train)
```

# 第六步：预测结果

```python
y_pred = classifier.predict(X_test)
y_pred
```

# 第七步：建立混淆矩阵

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

# 第八步：训练集结果的可视化

```python
from matplotlib.colors import ListedColormap
```

```python
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01))
plt.contourf(X1, X2, 
             classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha=0.75, 
             cmap=ListedColormap(('red', 'green')))

plt.xlim(X1.min(), X1.max()) # x轴的刻度范围
plt.ylim(X2.min(), X2.max()) # y轴的刻度范围

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set==j, 0], # x轴
                X_set[y_set==j, 1], # y轴
                c=ListedColormap(('red', 'green'))(i), 
                label=j)

plt.title('Random Forest Classification (Training set)') # 图标标题
plt.xlabel('Age') # x轴标签
plt.ylabel('Estimated Salary') # y轴标签 
plt.legend() # 图例
plt.show()
```

# 第九步：测试集结果的可视化

```python
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step=0.01))
plt.contourf(X1, X2, 
             classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha=0.75, 
             cmap=ListedColormap(('red', 'green')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set==j, 0], 
                X_set[y_set==j, 1],
                c=ListedColormap(('red', 'green'))(i), 
                label=j)

plt.title('Random Forest Classification (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
